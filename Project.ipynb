# merging

```python
#merging
#pd.merge(left,right,on="comman column",how='inner')
#left=first df
#right=second df
#on = consider comman column in 2 dataframes and merge on that column
import pandas as pd
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
registrations
```

```python
logins
```

```python
pd.merge(registrations,logins,on='name')  #how='inner' is bydefault
```

```python
pd.merge(registrations,logins,on="name",how='inner')
#how='inner' means it will take comman records in both tables 
```

```python
pd.merge(registrations,logins,on="name",how='left')
#how=left ,it will select all records from left(dataframe1) and comman records from right(dataframe2)
```

```python
pd.merge(registrations,logins,on="name",how='right')
#how=right ,it will select all records from right(dataframe2) and comman records from left(dataframe1)
```

```python
pd.merge(registrations,logins,on="name",how='outer')
#how= outer ,it will select all records from both tables
```

```python
registrations = registrations.set_index("name")
registrations
```

```python
logins
```

```python
#no comman column in both dataframes
pd.merge(registrations,logins,left_index=True,right_on='name')
```

```python
#boxplot
#box plot 
#Box Plot is the visual representation of the depicting groups of numerical data through their quartiles.
#Boxplot is also used for detect the outlier in data set. It captures the summary of the data efficiently with a simple box
#and whiskers and allows us to compare easily across groups. Boxplot summarizes a sample data using 25th, 
#50th and 75th percentiles. These percentiles are also known as the lower quartile, median and upper quartile.

#A box plot consist of 5 things.

#1.Minimum
#2.First Quartile Q1 or 25%(percentile)
#3.Median (Second Quartile) Q2 or 50% percentile
#4.Third Quartile Q3 or 75% percentile
#5.Maximum

#IQR=Inter quartile range = Q3-Q1
#min = Q1 - (1.5 * IQR)
#max = Q3 + (1.5 * IQR)

#distribution of total bill for a day
sns.boxplot(x="day",y="total_bill",data=df)
```

# Heatmap or matrix plot

```python
#Heatmap or matrix plot
#heatmap is a way of representing data in 2D form.the data values are represented as colors in the graph.the goal of heatmap
#is to provide a colored visual summary of information.
#corr() method will actually get rid of the columns that are not suitable for correlation for you.
import seaborn as sns
df=sns.load_dataset('tips')
df.head()
```

```python
#variables on columns and rows are must be same for the matrix plot(heatmap)
x=df.corr()
x
```

```python
sns.heatmap(x,annot=True)
```

```python
sns.heatmap(x,annot=True,cmap='RdYlBu')
```

```python
sns.heatmap(x,annot=True,cmap='viridis')
```

```python
sns.heatmap(x,annot=True,cmap='RdYlGn')
```
